```python
from pymongo import MongoClient
def get_database():
   with open(".env.development", "r") as f:
      lines = f.readlines()
      for line in lines:
         if "DB_CONNECTION_STRING" in line:
            DB_CONNECTION_STRING = "=".join(line.split("=")[1:]).strip()
            break
   client = MongoClient(DB_CONNECTION_STRING)
   db = client.reverb
   return db
db = get_database()
```

```python
posts_collection = db["posts"]
```

```python
# posts_collection.find()
item_details = posts_collection.find()
for item in item_details:
   # This does not give a very readable output
   print(item)
```
